## 2.2 파일 통합, data merge

### file directory setting

```python
outfile = './data/정제/merged_data.txt'
directory = './data/'
outfile = open(outfile,'w')
files = os.listdir(directory)
```

### file merge

```python
for filename in files:
    if ".txt" not in filename:
        continue
    print(filename+"통합")
    try:
        file = open(directory+filename, encoding='utf-8')
    except:
        file = open(directory+filename, encoding='cp949')
    for line in file:
        if line!="\n" and line!="":
            outfile.write(cleaning(line))
    outfile.write("\n\n")
    file.close
outfile.close()
        
```
